```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# Load diamond dataset into a pandas DataFrame 
df = sns.load_dataset('diamonds')

# Inspect first rows of the DataFrame
print(df.head())
```

```python
##Step Initial Data Inspection


print(df.head(10))
print(df.shape)
print(df.dtypes)
```

```python
#Step 3. Initial Descriptive Statistics 
print(df.describe())
```

```python
##Step 4. Initial Data Distribution for Numerical Columns
# Inspect histogram by numerical column
plt.figure(figsize=(8,6))
plt.hist(df["price"], 
        bins = 8,
        color = "#108A99" )
plt.title("Distribution of Diamond prices", fontsize = 12, weight = "bold")
plt.xlabel("Price")
plt.ylabel("Number of Diamonds" )
plt.show()
```

From above histogram, dataset contains information about 53,940 round-cut diamonds. diamond price ranges between \$500 to \$18,500. Number of diamonds are more with the price below \$2500 when compared with \$17500

## Step 5. Initial Data Distribution for Categorical Columns


```python
df['cut'].value_counts()
```

Fair: This value occurs 1,610 times.
Good: This value occurs 4,906 times.
Very Good: This value occurs 12,082 times.
Premium: This value occurs 13,791 times.
Ideal: This value occurs 21,551 times.

```python
# Inspect value counts for all categorical columns
for col in df.select_dtypes(include=['object', 'category']).columns:
    # Display count plot
    sns.countplot(x=col, data=df)
    plt.title(f'Distribution of {col}')
    plt.show()

# Show all plots
plt.show()
```


Ideal cut diamonds and SI1 clarity are very common, The most common color is G, which is in the middle of the distribution. There are more diamonds in the top half of the distribution than in the bottom.


## Rename price column and Insert column - weight 

```python
df = df.rename(columns={'price': 'Cost'})
import numpy as np
df['Weight'] = np.random.randint(0,5, size=len(df))
print(df.head(10))
```


### Goal 1: To determine how the cost varies based on carat and cut of diamonds. 
### scatter plot is used

```python
sns.scatterplot(
      data=df,   x='carat', y='Cost', hue = 'cut'
)
```

Ideal cut and premium cut diamonds are very common, as carat size increases the value of diamond price goes up. 

### Goal 2: To determine cost range based on color of diamonds. 
### box plot is used

```python
sns.boxplot( x=df["color"], y=df["Cost"] )
```

Cost range of diamonds with color I and J are very expensive compared to other colors.
