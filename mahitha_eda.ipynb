```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# Load diamond dataset into a pandas DataFrame 
df = sns.load_dataset('diamonds')

# Inspect first rows of the DataFrame
print(df.head())
```

```python
##Step Initial Data Inspection


print(df.head(10))
print(df.shape)
print(df.dtypes)
```

```python
#Step 3. Initial Descriptive Statistics 
print(df.describe())
```

```python
##Step 4. Initial Data Distribution for Numerical Columns
# Inspect histogram by numerical column
plt.figure(figsize=(8,6))
plt.hist(df["price"], 
        bins = 8,
        color = "#108A99" )
plt.title("Distribution of Diamond prices", fontsize = 12, weight = "bold")
plt.xlabel("Price")
plt.ylabel("Number of Diamonds" )
plt.show()
```

From above histogram, diamond price ranges from between around $500 to $18,500. Most of the diamonds falls in the lesser price.
